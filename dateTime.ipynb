```python
import datetime
from datetime import datetime
from dateutil.relativedelta import relativedelta
```

```python
Today=datetime.today().strftime("%Y-%m-%d")
```

```python
earningRelease=['3-31','5-15','8-14','11-14']
```

```python
# 把財報發布的日期換算成天數
def earnReleasePoint():
    content=[]
    for item in earningRelease:
        date=datetime.strptime(item,"%m-%d")
        point=date.month*30+date.day
        content.append(point)
    return content
    
print(earnReleasePoint())
```

```python
# 找出最近的財報發布日
def revenueRelease(today=Today):
    today=datetime.strptime(today,"%Y-%m-%d")
    if today.day>15:
        return today.replace(day=15).strftime("%Y-%m-%d")
    else:
        today.replace(month=today.month-1)
        return today.replace(day=15).strftime("%Y-%m-%d")
```

```python
def monthsBefore(n,today=Today):
    today=datetime.strptime(today,"%Y-%m-%d")
    return (today- relativedelta(months=n)).strftime("%Y-%m-%d")
```

```python
def yearsBefore(n,today=Today):
    today=datetime.strptime(today,"%Y-%m-%d")
    return (today- relativedelta(years=n)).strftime("%Y-%m-%d")
```

```python
# 對應第幾季
def checkSeason(today=Today):
    today=datetime.strptime(today,"%Y-%m-%d")
    checkPoint=today.month*30+today.day
    if checkPoint>=earnReleasePoint()[0] and checkPoint<earnReleasePoint()[1]:
        return 1
    elif checkPoint>=earnReleasePoint()[1] and checkPoint<earnReleasePoint()[2]:
        return 2
    elif checkPoint>=earnReleasePoint()[2] and checkPoint<earnReleasePoint()[3]:
        return 3
    else:
        return 4
```

```python
def earningReleaseDate(today=Today):
#     today=datetime.strptime(today,"%Y-%m-%d")
    content=[]
    earningRelease=['3-31','5-15','8-14','11-14']
    for i in range(len(earningRelease)):
        if i>=checkSeason(today):
            earningReleaseConvert=str(datetime.today().year-1)+"-"+earningRelease[i]
            content.append(earningReleaseConvert)
    for i in range(len(earningRelease)):
        if i<checkSeason(today):
            earningReleaseConvert=str(datetime.today().year)+"-"+earningRelease[i]
            content.append(earningReleaseConvert)
    return content
```
