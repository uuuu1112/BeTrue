```python
import pandas as pd
from datetime import *
import requests
from Stocker.stocker import Stocker
```

```python
today=datetime.now()
today=datetime.strftime(today,'%Y-%m-%d')
print(today)
```

```python
def indexPredice(type):
  url = "https://api.finmindtrade.com/api/v4/data"
  parameter = {
      "dataset": "TaiwanStockPrice",
      "data_id": type,
      "start_date": "2000-01-01",
      "end_date": today,
      "token": "", # 參考登入，獲取金鑰
  }
  resp = requests.get(url, params=parameter)
  data = resp.json()
  data = pd.DataFrame(data["data"])
  data['date']=data['date'].map(lambda x:datetime.strptime(x,"%Y-%m-%d"))
  data.set_index('date',inplace=True)
  df=data[['close']]
  price=df.squeeze()
  return price
```

```python
def indexMap():
  url = "https://api.finmindtrade.com/api/v4/data"
  parameter = {
      "dataset": "TaiwanStockInfo",
      "token": "", # 參考登入，獲取金鑰
  }
  resp = requests.get(url, params=parameter)
  data = resp.json()
  data = pd.DataFrame(data["data"])
  index=data['industry_category']=='Index'
  index=data[index]
  index=index[['stock_id','stock_name']]
  index.set_index('stock_id',inplace=True)
  indexMap=index.to_dict()
  indexMap=indexMap['stock_name']
  return indexMap
```

```python
print(indexMap())
```

```python
days=250
```

```python
for type in indexMap():
  print(type,indexMap()[type])
  target = Stocker(indexPredice(type))
  model, model_data = target.create_prophet_model(days=days)
```
